### Importing Necessary Libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('../../Data_After_Cleaning/HospInfo_Cleaned.csv')
```

### Data Understanding

### Data Cleaning

```python
def cleaned_data(df):
    # droping the county name missing data
    df.dropna(subset=['County Name'], inplace=True) 
    #handling missing values
    columns_to_fill = ['Hospital overall rating footnote',
          'Mortality national comparison footnote',
          'Safety of care national comparison footnote',
          'Readmission national comparison footnote',
          'Patient experience national comparison footnote',
          'Effectiveness of care national comparison footnote',
          'Timeliness of care national comparison footnote',
          'Efficient use of medical imaging national comparison footnote']
    df[columns_to_fill] = df[columns_to_fill].fillna('Unknown',inplace=False)

    df['Hospital overall rating'] = pd.to_numeric(df['Hospital overall rating'], errors='coerce')

    state_ratings = df.groupby('State')['Hospital overall rating'].mean().dropna().sort_values(ascending=False)

    # Top 5 and Bottom 5 states
    top_states = state_ratings.head(5)
    bottom_states = state_ratings.tail(5)
    combined_states = pd.concat([top_states, bottom_states])

    return df
```

### Exploratory Data Analysis

```python
def visualizations(df):
    # Count the number of each hospital type per state
    hospital_type_distribution = df.groupby(['State', 'Hospital Type']).size().unstack(fill_value=0)

    # Plot a stacked bar chart of hospital types for each state
    plt.figure(figsize=(8, 5))
    hospital_type_distribution.plot(kind='bar', stacked=True, figsize=(16, 8), colormap='tab20')
    plt.title('Distribution of Hospital Types Across States')
    plt.ylabel('Number of Hospitals')
    plt.xlabel('State')
    plt.legend(title='Hospital Type', bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.tight_layout()
    plt.show()

    # Convert 'Emergency Services' column from boolean to 0/1 integer
    df["emergency_available"] = df["Emergency Services"].astype(int)

    # Group by state and calculate fraction of hospitals offering emergency services
    state_emergency = (
    df.groupby("State")["emergency_available"]
      .agg(availability_rate="mean",  
           total_hospitals="count")
      .sort_values("availability_rate", ascending=False)
      .round(3))

    # Plot emergency availability per state
    plt.figure(figsize=(9, 5))
    state_emergency["availability_rate"].plot(kind="bar",
                                              color="seagreen",
                                              edgecolor="black")
    plt.title("Emergency‑Service Availability Rate by State")
    plt.xlabel("State")
    plt.ylabel("Availability Rate  (fraction of hospitals with emergency dept.)")
    plt.xticks(rotation=90)
    plt.ylim(0, 1.05)
    plt.tight_layout()
    plt.show()

    # Convert rating to numeric 
    df['Hospital overall rating'] = pd.to_numeric(df['Hospital overall rating'], errors='coerce')

    # Calculate average rating per state
    state_ratings = df.groupby('State')['Hospital overall rating'].mean().dropna().sort_values(ascending=False)

    # Top 5 and Bottom 5 states
    top_states = state_ratings.head(5)
    bottom_states = state_ratings.tail(5)
    combined_states = pd.concat([top_states, bottom_states])

    # Plot combined bar chart 
    plt.figure(figsize=(10, 5))
    colors = ['green' if state in top_states.index else 'red' for state in combined_states.index]
    combined_states.sort_values().plot(kind='barh', color=colors)
    plt.title('Top 5 and Bottom 5 States by Average Hospital Rating')
    plt.xlabel('Average Hospital Rating')
    plt.ylabel('State')
    plt.tight_layout()
    plt.show()

    # Subset necessary columns
    df_satisfaction = df[['Hospital Type', 'Patient experience national comparison']]

    # Remove rows where satisfaction info is not available
    df_satisfaction = df_satisfaction[df_satisfaction['Patient experience national comparison'] != 'Not Available']

    # Count satisfaction levels for each hospital type
    satisfaction_counts = df_satisfaction.groupby(['Hospital Type', 'Patient experience national comparison']).size().unstack(fill_value=0)

    # Convert to percentage for each hospital type
    satisfaction_percentages = satisfaction_counts.div(satisfaction_counts.sum(axis=1), axis=0) * 100

    # Plot stacked bar chart
    plt.figure(figsize=(11, 5))
    satisfaction_percentages.plot(kind='bar', stacked=True, figsize=(12, 7), colormap='viridis')
    plt.title('Patient Satisfaction by Hospital Type')
    plt.ylabel('Percentage of Hospitals')
    plt.xlabel('Hospital Type')
    plt.legend(title='Patient Experience Comparison', bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.tight_layout()
    plt.show()

    # Hospital Ratings Distribution by Hospital Type
    plt.figure(figsize=(12, 6))
    sns.boxplot(data=df, x='Hospital Type', y='Hospital overall rating', palette='Set3')
    plt.title('Distribution of Hospital Ratings by Hospital Type')
    plt.xlabel('Hospital Type')
    plt.ylabel('Hospital Overall Rating')
    plt.show()

    # Correlation Heatmap for Key Numeric Variables
    plt.figure(figsize=(8, 6))
    numeric_df = df[['Hospital overall rating', 'emergency_available']]
    sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', center=0)
    plt.title('Correlation Heatmap')
    plt.show()

    # Hospital Count by Ownership Type Across States
    ownership_distribution = df.groupby(['State', 'Hospital Ownership']).size().unstack(fill_value=0)

    plt.figure(figsize=(16, 8))
    ownership_distribution.plot(kind='bar', stacked=True, colormap='Paired')
    plt.title('Distribution of Hospital Ownership Types Across States')
    plt.ylabel('Number of Hospitals')
    plt.xlabel('State')
    plt.legend(title='Hospital Ownership', bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.tight_layout()
    plt.show()

    # Hospital Count per State
    plt.figure(figsize=(12, 6))
    df['State'].value_counts().plot(kind='bar', color='skyblue', edgecolor='black')
    plt.title('Number of Hospitals per State')
    plt.xlabel('State')
    plt.ylabel('Number of Hospitals')
    plt.tight_layout()
    plt.show()


```

```python
visualizations(df)
```
